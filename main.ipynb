# Data Visualization 

```python
import matplotlib.pyplot as plt

sorted_data = data_without_null.sort_values(by='ANNUAL', ascending=False)

plt.figure(figsize=(10, 10)) 
plt.barh(sorted_data['SUBDIVISION'], sorted_data['ANNUAL'])
plt.xlabel('Annual Rainfall (in mm)')
plt.ylabel('SUBDIVISION of States')
plt.title('Annual Rainfall by Subdivision (Sorted)')
plt.grid(axis='x')
plt.show()
```

```python
plt.figure(figsize=(15,8))
data_without_null.groupby("YEAR").sum()['ANNUAL'].plot(kind="line",color='r',marker='*')
plt.xlabel("YEARS",size=12)
plt.ylabel("RAINFALL IN MM",size=12)
plt.grid(axis="both")
plt.title("Rainfall over Years")
plt.show()
```

```python
data_without_null[['YEAR','JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].groupby("YEAR").sum().plot(kind="line",figsize=(18,8))
plt.xlabel("Year",size=13)
plt.ylabel("Rainfall in MM",size=13)
plt.title("Year V/S Rainfall in each month",size=20)
plt.show()
```

```python
data_without_null[['YEAR','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']].groupby("YEAR").sum().plot(kind="line",figsize=(18,8))
plt.xlabel("Year",size=13)
plt.ylabel("Rainfall in MM",size=13)
plt.title("Year V/S Rainfall in range of months",size=20)
plt.show()
```

```python
data_without_null[['SUBDIVISION','JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].groupby("SUBDIVISION").sum().plot(kind="barh",stacked=True,figsize=(25,15))
plt.xlabel("Rainfall in MM",size=15)
plt.ylabel("Sub-division",size=15)
plt.title("Sub-division V/S Rainfall in each month",size=20)
plt.grid(axis="x",linestyle="-")
plt.show()
```

```python
data_without_null[['SUBDIVISION','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']].groupby("SUBDIVISION").sum().plot(kind="barh",stacked=True,figsize=(25,15))
plt.xlabel("Rainfall in MM",size=15)
plt.ylabel("Sub-division",size=15)
plt.title("Sub-division V/S Rainfall in each month",size=20)
plt.grid(axis="x",linestyle="-")
plt.show()
```
